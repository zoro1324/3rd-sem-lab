Question : Data Cleansing and Preparation using Python

Analyze the network traffic dataset to detect missing values, incorrect data types, and duplicate entries. Apply suitable data cleaning techniques to ensure structural accuracy and consistency.
Tasks:
1.	Load and inspect the dataset using pandas.
2.	Identify and treat missing data using mean and mode imputation.
3.	Convert Timestamp to datetime format and remove duplicates.
4.	Summarize the cleaning process.


```python
import pandas as pd
```

```python
#Hide inline suggestion I am trying to learn pandas
df = pd.read_csv(r'datasets\data_1.csv')
df.info()
```

```python
mean_packets = df['Packet_Size'].mean()
print("Mean Packet Size:", mean_packets)
mode_protocol = df['Protocol'].mode()[0]
print("Mode Protocol:", mode_protocol)
```

```python
df['Packet_Size'] = df['Packet_Size'].fillna(mean_packets)
df['Protocol'] = df['Protocol'].fillna(mode_protocol)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```
