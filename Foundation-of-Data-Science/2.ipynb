Exploratory Visualization using Python
Visualize web session data to uncover relationships between Session Duration, Bounce Rate, and Page Views. Identify user behavior trends using graphical analysis.
Tasks:
1.	Create scatter and box plots using matplotlib or seaborn.
2.	Compare engagement patterns across traffic sources.
3.	Add labels, legends, and titles for clarity.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/data_2.csv')
df.info()
```

```python
sns.scatterplot(data=df,x='Session Duration',y='Bounce Rate')
plt.title('Session Duration vs Bounce Rate')
plt.xlabel('Session Duration (seconds)')
plt.ylabel('Bounce Rate (%)')
plt.legend()
plt.show()
```

```python
# 2. Scatter Plot  Relationship between Session Duration and Page Views
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Session Duration', y='Page Views')
plt.title("Session Duration vs Page Views by Traffic Source")
plt.xlabel("Session Duration (seconds)")
plt.ylabel("Page Views")
plt.legend()
plt.show()
```

```python
sns.boxplot(data=df)
plt.title("Distribution of Web Session Metrics")
plt.ylabel("Value")
plt.show()
```
